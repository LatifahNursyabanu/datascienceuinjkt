### Common List Operation

```python
# Modifying element 
# You can change the value of any element in a list if you know the position of that item.

dogs = ['border collie', 'australian cattle dog', 'labrador retriever']

dogs[0] = 'australian shepherd'
print(dogs)
```

## Practice makes perfect!

Buat function sehingga komputer dapat mengonversi dari input detik ke menit dan jam

```python
def convert_seconds(second):
  hours = second // 3600
  minutes = second % 3600 // 60
  seconds = second - ((hours * 3600) + (minutes * 60)) 
  return hours, minutes, seconds

convert_seconds(5000)
```

```python
def square(n):
    return n*n

def sum_squares(x):
    sum = 0
    for n in range(x):
        sum += n*n
    return sum

print(sum_squares(5)) # Should be 30 (1+4+9+16)
```

```python
def factorial(n):
    result = 1
    for i in range(1,n+1):
        result = result*i
    return result

print(factorial(4)) # should return 24
print(factorial(5)) # should return 120
```

```python
# Nested Loop

for i in range(3):
    for j in range(2):
        print(i,j)
```

```python
def is_power_of_two(n):
  # Check if the number can be divided by two without a remainder
  if n != 0:
    while n % 2 == 0:
      n = n / 2
  # If after dividing by two the number is 1, it's a power of two
  if n == 1:
    return True
  return False
  

print(is_power_of_two(0)) # Should be False
print(is_power_of_two(1)) # Should be True
print(is_power_of_two(8)) # Should be True
print(is_power_of_two(9)) # Should be False
```

# Break and Continue

You can interrupt both while and for loops using the break keyword. We normally do this to interrupt a cycle due to a separate condition.

You can use the continue keyword to skip the current iteration and continue with the next one. This is typically used to jump ahead when some of the elements of the sequence arent relevant.

```python
fruits = ['jeruk','apel','mangga','pisang']
for i in fruits:
    if i=='apel':
        break
    print(i)
```

```python
fruits = ['jeruk','apel','mangga','pisang']
for i in fruits:
    if i=='apel':
        continue
    print(i)
```
