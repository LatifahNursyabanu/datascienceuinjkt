# Matrix

```python
# Matrix manipulation

array2D = np.array([[1, 2, 3], 
                    [3, 4, 5], 
                    [5, 6, 7]])

array2D[0, 2] = 100
array2D
```

```python
# 3 ways of transpose syntax

array2D.transpose()

# np.transpose(array)
# array.T
```

```python
# Flatten matriks

array2D = np.array([[1, 2, 3], 
                    [3, 4, 5], 
                    [5, 6, 7]])
array2D.ravel()
```

```python
# Basic aggregate functions

array2D = np.array([ [[1, 2], [2, 3]], 
                     [[3, 4], [4, 5]] ])

print("Mean value is: ", array2D.mean())
print("Max value is: ", array2D.max())
print("Min value is: ", array2D.min())
print("Min value is: ", array2D.sum())
```

```python
# tile dan repeat

array2D = np.array([ [[1, 2], [2, 3]],
                    [[3, 4], [4, 5]] ])

print("tile: ", np.tile(array2D,2))
print("repeat: ", np.repeat(array2D,2))
```

```python
# unique and count

array2D = np.array([ [[1, 2], [2, 3]],
                    [[3, 4], [4, 5]] ])

unik, count = np.unique(array2D, return_counts=True)
print("Unik: ", unik)
print("Count: ", count)
```

# Sigmoid
image.png


```python
import numpy as np

# 1. Sigmoid

def sigmoid(x):
    sigmoid = np.exp(x) / (1 + np.exp(x))
    return sigmoid

print('Sigmoid of 5 is ' + str(sigmoid(5)))
```
